````python
address_list = []

def address_info(addresses):
    global address_list
    
    for i in range(len(addresses)):
        address_dict = {}
        try:
            d = json.loads(addresses[i])
            address_dict['Address'] = d["RequestAddress"]['AddressLine']
            address_dict["Longitude"] = d["SuggestedAddress"][0]['Address']['PremisesAddress']['GeospatialInformation'][0]["Longitude"]
            address_dict['Latitude'] = d["SuggestedAddress"][0]['Address']['PremisesAddress']['GeospatialInformation'][0]['Latitude']
        
        except:
            address_dict['Address'] = d["RequestAddress"]['AddressLine']
            address_dict["Longitude"] = 'np.nan'
            address_dict['Latitude'] = 'np.nan'

        address_list.append(address_dict)
        
    return address_list


address_info(addresses)[10]

````
