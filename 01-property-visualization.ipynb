# Property Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
```

```python
df = pd.read_csv('../data/Hong_Kong_Clean.csv')
```

```python
df.head()
```

```python
df['date1'] = pd.to_datetime(df['date'])
# df.head()
df['year'] = df['date'].apply(lambda x:x[0:4])
df.head()

```

**Plot a graph showing the price per square feet over the years.**

```python
fig, ax = plt.subplots(figsize=(13,8))
```

```python
fig, ax = plt.subplots(figsize=(13,8))
sns.barplot(data=df, x = 'year',y ="sqft_price" ,palette="pastel" )
```

**Plot a graph showing the area per square feet against each district.**

```python
plt.figure(figsize=(8,20))
```

```python
plt.figure(figsize=(8,20))
sns.barplot(data=df, x = 'sqft_area',y ="district" ,palette="pastel" )
```

**Plot a graph showing relationship between floor and the price**

```python
plt.figure(figsize=(13,8))
```

```python
df['price_value2']=df['price_value']/1000000
plt.figure(figsize=(13,8))
sns.barplot(data=df, x = 'floor',y ="price_value2" ,palette="pastel" )
# plt.ylim(0, 10)
```
