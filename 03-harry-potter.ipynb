````python
with open("../data/PrisonerOfAzkaban.txt","r") as f:
    text = f.read()
    chapters = text.split("CHAPTER")
    # type(chapters) = list
    # each chapter --> become a string
    for x in range(len(chapters)):
        # chap_str = " "
        # chap_str_1 =chap_str.join(chapters[x])
        chap_sent = len(chapters[x].split('.'))
        chap_word = len(chapters[x].split())
        print(f'The number of sentences of chapter {x+1}: {chap_sent}')
        print(f'The number of words of chapter {x+1}:{chap_word}')
    
````

````python
dict_list=[]
# REMARK: dict_list[0] = chapter 1 dictionary!!!!

with open("../data/PrisonerOfAzkaban.txt","r") as f:
    text = f.read()
    chapters = text.split("CHAPTER")
    for x in range(1,len(chapters)):
        word_in_chap = chapters[x].split()
        words_chap_dict = {}
        for i in word_in_chap: 
            if i not in words_chap_dict:
                words_chap_dict [i]=1
            else:
                words_chap_dict [i]+=1
        dict_list.append(words_chap_dict)
        


    
# # print(dict_list[0])
# chap1_dict = dict_list[0]
# print(chap1_dict)
   
people = ['Harry','Ron','Voldemort','Sirius','Vernon','Hermione']
school = ['Azkaban', 'Hogwarts','Gryffindor' , 'Hufflepuff' ,'Ravenclaw', 'Slytherin']
ppl_count_chap_list = []
school_count_chap_list = []


for j in range(len(dict_list)):
    ppl_count_chap = {}
    for i in range(len(people)):
        try:
            ppl_count_chap[people[i]] = dict_list[j][people[i]]
        except:
            ppl_count_chap[people[i]] = 0
    ppl_count_chap_list.append(ppl_count_chap)
    #print(f'The people count in chanper{j+1} is {ppl_count_chap}')
# print(ppl_count_chap_list)
for j in range(len(dict_list)):
    school_count_chap = {}
    for i in range(len(school)):
        try:
            school_count_chap[school[i]] = dict_list[j][school[i]]
        except:
            school_count_chap[school[i]] = 0
    school_count_chap_list.append(school_count_chap)
    print(f'The place count in chanper{j+1} is {school_count_chap}')



# print(dict_list[1]['Harry'])
# print(ppl_count_chap)

    
````
