** What is the average Purchase Price? **

```python
ecom['Purchase Price'].mean()
```

```python
# average purchase price per credit card
ecom.groupby('Credit Card')['Purchase Price'].mean()
```

```python
# create a new column called credit card mean - which will have the mean credit card value of corresponding credit card


ecom[ecom['Credit Card']==675957666125]['Purchase Price'].mean()
```

```python
# ANS
def engineer_card_mean(x):
    return ecom[ecom['Credit Card']== x]['Purchase Price'].mean()

ecom['Card Mean'] = ecom['Credit Card'].apply(engineer_card_mean)
```

```python
ecom = ecom.drop(columns='Credit Card Mean')
```

** Hard: How many people have a credit card that expires in 2025? **

```python
ecom['CC Exp Date'].str.contains('/25').value_counts()
```

```python
# Method 2

ecom["_2025"]=ecom["CC Exp Date"].apply(lambda x: "True" if x[-2:]=="25" else "False")

ecom["_2025"].value_counts()

# In 1 Line
ecom["CC Exp Date"].apply(lambda x: "True" if x[-2:]=="25" else "False").value_counts()
```
