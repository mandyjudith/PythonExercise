```python
# MY TRIAL 1
#  df.groupby('occupation')['gender'].value_counts()

x = df.groupby(['occupation','gender'])['gender'].count()
y = df.groupby(['occupation'])['gender'].count()
ratio = ((x/y)*100).round(2)
print(ratio) 
```

```python
# MY ANS
male_ratio = df.pivot_table(index='occupation',columns='gender',aggfunc= 'size')
# print(male_ratio)

sums = male_ratio[['F','M']].sum(axis=1)
r = (male_ratio['M']/sums)*100
s = r.sort_values(ascending=False)
print(s)

```

 ** 6.) ** what is the maximun and minimum age for each occupation?

```python
# df_max = df.groupby(df['occupation'])['age'].max()
# df_min = df.groupby(df['occupation'])['age'].min()

z = df.groupby('occupation')['age'].describe().loc[:,['max','min']]
z
```

```python
# Method 2 : merge()
min_max = pd.DataFrame(df.groupby(['occupation'])["age"].max()).merge(pd.DataFrame(df.groupby(["occupation"])["age"].min()), on="occupation", how="inner")

min_max= min_max.rename(columns={'age_x':'max','age_y':'min'})
min_max
```

```python
# Method 3 : merge()
df_max = df.groupby(['occupation'])["age"].max()
df_min = df.groupby(['occupation'])["age"].min()

combine = pd.merge(df_max,df_min, on="occupation")
combine =  combine.rename(columns= {'age_x':'max','age_y':'min'})
combine
```
