**13. How to find common values between two arrays? **

```python
# %load solutions/np_13.py
```

```python
x = np.arange(0,5)
y = np.arange(3,8)
z = ([])
for i in x:
    for j in y:
        if i == j:
            z.append(i)
print(z)

import matplotlib.pyplot as plt
import matplotlib as mpl

plt.plot(x,y,z)
```

**14. How to get the dates of yesterday, today and tomorrow? **

```python
yesterday = np.datetime64('today', 'D') - np.timedelta64(1, 'D')
print("Yestraday: ",yesterday)
today     = np.datetime64('today', 'D')
print("Today: ",today)
tomorrow  = np.datetime64('today', 'D') + np.timedelta64(1, 'D')
print("Tomorrow: ",tomorrow)
```

**20. Find the nearest value from a given value in an array**

```python
# %load solutions/np_20.py
```

```python
def nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array-value)).argmin()
    return array[idx]

array = np.random.randint(1,20,8)
print(array)
value = 11
print(nearest(array,value))
```

**21. How to get the diagonal of a dot product Of A and A transpose? ** You can get A transpose by using A.T



```python
# %load solutions/np_21.py
```

```python
x = np.arange(9).reshape((3,3))
np.diag(x)
np.diag(x, k=1)
np.diag(x, k=-1)
y = np.diag(np.diag(x))
print(y)
np.dot(y,y.T)
```

```python
A = np.random.uniform(0,1,(5,5))
print(np.dot(A,A))

np.diag(A @ A)
# A@A = np.dot(A,A)
```
